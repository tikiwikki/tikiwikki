Background information about Phishing:

Phishing is a cyber-attack which targets naive online users tricking into revealing sensitive information such as username, password, social security number or credit card number etc. Attackers fool the Internet users by masking webpage as a trustworthy or legitimate page to retrieve personal information. 

There are many anti-phishing solutions such as **blacklist or whitelist**, heuristic and visual similarity-based methods proposed to date, but online users are still getting trapped into revealing sensitive information in phishing websites. In this paper, we propose a novel classification model, based on heuristic features that are extracted from URL, source code, and third-party services to overcome the disadvantages of existing anti-phishing techniques.

There are 4 groups in the documentation:

**1. Address bar based features (index 0 to 11)** following data and explaination:

    0. having_ip_address = Using the IP address

    1. url_length = Long URL to hide the suspicious part

    2. shortining_service = using URL shortnening services "TinyURL" e.g. bitly/19DXSk4"

    3. having_at_symbol = URL has a "@" symbol

    4. double_slash_redirecting = Redirecting using "//"

    5. prefix_suffix = Adding Prefix or SUffix separated by (') to the Domain

    6. having_sub_domain = = Sub Domain and Multi Sub Domains

    7. sslfinal_state = HTTP (Hyper Text Transfer Protocol with Secure Sockets Layer) - check the trust certificate issues and certificate age (< 2)

    8. domain_registration_length = Domain registration Length - phishing website is shortlived so < 1 year could be phishing

    9. favicon = favicon: graphic icon on the top, to identify the webpage or if loaded from another domain to try to mimic the trustworthy

    10. port = Using Non-Standard Port when preferred status of port is continious open sounts phishy

    11. https_token = Existence of HTTPS token in the Domain part of the URL e.g. http://**https**-www.-paypal......
   
**2. Abnormal based features (12 to 17)**


    12. Request_URL = contain external objects within a webpage e.g. images - to mimic  the trustworthy website

    13. URL_of_Anchor = elements with a <a> tage or any other tags

    14. links_in_tags = Links in '<'Meta'>', '<'Script'>' and '<'Link'>' tags 

    15. sfh = Server Form Handler (SFH), if this contains an empty string or "about :blank" looks suspicious

    16. submitting_to_email = web form with a function like "mail()" or "mailto:" looks very phishy.

    17. abnormal_url = This is from a a data base "WHOIS"


**3. HTML & java script based features (18 to 22)**

    18. redirect - website forwarding/redirecting to another website very suspicious. THis can happen but not often

    19. on_mouseover = this functin is used to show the fake URL in the status bar.

    20. rightclick = disabling right click option
  
    21. popupwindow = Using a Pop-up window to submit personal information which does not happen in a normal website 
    
    22. iframe = IFrame redirection : a HTML tag to display an additional website to mimic and the fishing website
  
**4. Domain based features (23 to 29)**

    23. age_of_domain = this information comes from WHOIS database minimal age for a phishing website is < 6 months
    
    24. dnsrecord = not recognize by the WOIS database

    25. web_traffic = number of visitors and the number of pages they visit (also the Alexa the Web Information company)
    
    26. page_rank = when there is a very low ranking
    
    27 google_index = if this website is not in the google index looks very phishy
    
    28. links_pointing_to_page = there are no other domains linking to them
    
    29. statistical_report = external information from other parties like PhishTank, StopBadware.
    

30. Result = -1 = legit 1 = phishy

```python
from scipy.stats import chi2_contingency
# Chi-square to identify features which have a low p-value
# Example threshold for significance
threshold = 0.05  # Retain features with p-value < 0.05

X = df.drop(columns=["Result"])  # Drop the target column to keep only the features

selected_features_chi = []

for col in X.columns:
    contingency_table = pd.crosstab(df[col], df['Result'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    if p < threshold:
        selected_features_chi.append(col)

print("Selected Features:", selected_features_chi)
```

**Learning Curve**

```python
# prepare for the learning curve for SVM
from sklearn.model_selection import learning_curve
from sklearn.svm import SVC

model = SVC(kernel='linear', class_weight='balanced', random_state=1701)

train_sizes, train_scores, valid_scores = learning_curve(
    model, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 10), scoring='accuracy')


```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
valid_mean = np.mean(valid_scores, axis=1)
valid_std = np.std(valid_scores, axis=1)
```

```python
# show the learning curve for SVM
plt.figure(figsize=(8, 6))

# Plot training scores
plt.plot(train_sizes, train_mean, label='Training Score', color='blue')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.2)

# Plot validation scores
plt.plot(train_sizes, valid_mean, label='Validation Score', color='orange')
plt.fill_between(train_sizes, valid_mean - valid_std, valid_mean + valid_std, color='orange', alpha=0.2)

# Add labels and legend
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve for SVM')
plt.legend(loc='lower right')
plt.show()
```
